## 1. Contexte général

La conduite automobile est une activité complexe qui mobilise plusieurs ressources cognitives : perception visuelle, attention soutenue, prise de décision, anticipation, contrôle moteur et gestion des imprévus.

Le papier CL-Drive s'intéresse donc à une question centrale :

> Peut-on estimer automatiquement la charge cognitive d'un conducteur à partir de signaux physiologiques et comportementaux mesurés pendant la conduite ?

#### Q1. Pourquoi est-il intéressant d'estimer la charge cognitive d'un conducteur ?

**Réponse :**

Charge cognitive trop importante augmente le risque d'accident.
Avec voitures intelligentes on pourrait alerter l'utilisateur de sa charge cognitive pour prévenir des accidents.

## 2. Définition : charge cognitive

#### Q2. Comment définir la charge cognitive ?

**Réponse :**

La charge cognitive c'est la concentration nécessaire à notre cerveau pour exécuter une ou plusieurs tâches.
Plusieurs tâches à exécuter en parallèle ou une seule tâche complexe augmente significativement la charge cognitive, sur une longue durée, cela peut entraîner une surcharge cognitive.

#### Q3. Dans le contexte de la conduite, de quels facteurs la charge cognitive peut-elle dépendre ?

**Réponse :**

Dans la conduite, la charge cognitive dépend de ce qu'il se passe sur la route (traffic élevé, route droite, changement de vitesse, tournant etc...) mais aussi de ce qu'il se passe au sein du véhicule ou tout simplement dans notre tête. Si on regarde notre téléphone, discute avec un passager, cela augmente la charge cognitive aussi.

Dans le papier, la charge cognitive est évaluée à l'aide d'une échelle subjective appelée **PAAS**, allant de 1 à 9.

 ## 5. Modalités mesurées

Le dataset CL-Drive est **multimodal**. Il contient plusieurs types de signaux enregistrés pendant la même tâche de conduite.

#### Q7. Compléter le tableau suivant.

| Modalité | Signification          | Signal mesuré                | Rôle possible      |
|---|------------------------|------------------------------|--------------------|
| EEG | ElectroEncéphaloGramme | Activité cérébrale           | Activité cérébrale |
| ECG | ElectroCardioGramme    | Battements cardiaques        | Stress             |
| EDA | ElectroDermalActivity  | Sudation                     | Stress, arousal    |
| Gaze | Regard                 | Mouvement des yeux, pupilles | Attention visuelle |

Dans le projet, la modalité principale sera l’**EEG**. Les autres modalités serviront éventuellement d’extensions pour discuter de la fusion multimodale.

#### Q8. Pourquoi peut-il être intéressant de combiner EEG, ECG, EDA et Gaze pour estimer la charge cognitive ?

**Réponse :**

Une mesure seule est moins fiable que plusieurs mesures, notamment pour le stress. Elles permettent de corriger entre elles les artéfacts qu'il peut y avoir.

## 7. Protocole expérimental

#### Q11. Quelle est la différence entre un **protocole expérimental** et un **dataset exploitable en machine learning** ?

**Réponse :**

Le protocole -> comment les données vont être collectées

Dataset exploitable -> résultat après traitement des données brut

#### Q12. Donner une description détaillée du protocole expérimental utilisé.

**Réponse :**

23 participants qui ont de l'expérience de conduite

Un simulateur avec 5 écrans, un volant, un système de simulation de vibration. Le tout pour répliquer au mieux les sensations réelles de conduite.

9 scénarios de 3 minutes, plus une phase d'adaptation au simulateur au début, progressivement plus complexes dans le but d'induire plus de charge cognitive.

4 capteurs, EEG, ECG, EDA, Gaze

#### Q13. Pourquoi les auteurs collectent-ils une baseline avant les scénarios ?

**Réponse :**

Parce qu'on est tous différents donc il faut une base pour avoir un delta. Mon coeur bat peut-être 2 fois plus vite que celui d'un athlète au repos.

#### Q14. Pourquoi prévoir un scénario d’orientation avant les scénarios principaux ?

**Réponse :**

Pour vérifier si l'utilisateur est à l'aise dans le simulateur, cela réduit les biais liés au simulateur et permet d'avoir une baseline.

#### Q15. Pourquoi prévoir des périodes de repos entre les scénarios ?

### Réponse

Pour retourner au niveau de base de charge cognitive. En plus le simulateur peut entraîner un inconfort si utilisé pendant trop longtemps

#### Q16. Pourquoi demander un score de charge cognitive toutes les 10 secondes plutôt qu’une seule fois à la fin du scénario ?

### Réponse

Parce que le modèle que l'on va utiliser va vérifier la qualité de ses réponses en prenant comme "ground truth" les valeurs données par les participants à ces moments là. Si un participant donne son score de charge cognitive après la simulation cela ne reflètera peut-être pas aussi bien la réalité.

## 8. Scénarios de conduite

Le dataset contient 9 scénarios principaux, conçus pour induire différents niveaux de charge cognitive.

#### Q17. Compléter le tableau suivant.

| Scénario | Description synthétique |
|---|---|
| 1 | Highway driving 80km/h |
| 2 | Night time driving 80km/h |
| 3 | Night time driving in the highway w/ snow 80km/h |
| 4 | Tennis ball challenge: Hit the tennis ball with the tire, maintain accuracy, try to accelerate |
| 5 | Slalom challenge: Navigate through gates, maintain accuracy, try to accelerate |
| 6 | Narrow passage challenge: Navigate through gates, maintain accuracy, try to accelerate |
| 7 | 90-degree turn challenge: Take 90 degrees left and right turns, maintain accuracy, try to accelerate |
| 8 | 3-point turn challenge: Drive while following instructions, take 3 point turn |
| 9 | Narrow alley challenge: Navigate through narrow alley, maintain accuracy, try to accelerate |

L'hypothèse générale est que les scénarios les plus complexes induisent une charge cognitive plus élevée.

#### Q18. Pourquoi utiliser plusieurs scénarios de conduite au lieu d'un seul scénario ?

**Réponse :**

Ca permet de mieux séparer les réponses aux stimuli, de permettre des pauses, de donner un sens logique aux transitions (sans pause entre les scénarios la voiture se téléporterait à une nouvelle posiiton à un autre moment sans transition ?)

#### Q19. Pourquoi peut-on s’attendre à une charge cognitive plus élevée dans les scénarios 8 et 9 ?

### Réponse

Parce qu'ils ont été conçus dans ce but. Conduire sur une autoroute de jour et dans une petite allée avec plein de tournants et de traffic ne demande absolument pas la même concentration dans la vraie vie.

#### Q20. La complexité du scénario suffit-elle à définir le label de charge cognitive ?

### Réponse

Non, on ne réagit pas tous pareil aux différents scénarios, c'est plutôt la moyenne des ressentis de tous les participants qui pourrait nous donner une idée du label à définir. Et de toute façon le label est demandé toutes les 10s et un scénario fait 3 minutes. Donc un scénario n'a pas de label, il a une série de labels qui peuvent varier au cours du temps.

## 9. Échantillonnage et segmentation temporelle

Les signaux EEG sont continus. Pour entraîner un modèle, il faut transformer ce signal continu en exemples individuels.

Dans CL-Drive, les scores PAAS de charge cognitive sont collectés toutes les 10 secondes.

Il est donc naturel de découper les signaux physiologiques en fenêtres de 10 secondes :

$$
X_i = \text{segment de 10 s}
$$

$$
y_i = \text{score PAAS associé}
$$

On obtient ainsi le problème suivant :

Le principe d’association segment-label est le suivant :

| Temps | Segment EEG | Score PAAS |
|---|---|---|
| 0–10 s | segment 1 | score 1 |
| 10–20 s | segment 2 | score 2 |
| 20–30 s | segment 3 | score 3 |

## 10. Formulation du problème d'apprentissage automatique

À partir du dataset CL-Drive, on peut formuler plusieurs problèmes de classification.

### Classification binaire

| Scores PAAS | Classe |
|---|---|
| 1 à 4 | charge faible |
| 5 à 9 | charge élevée |

### Classification ternaire

| Scores PAAS | Classe |
|---|---|
| 1 à 3 | charge faible |
| 4 à 6 | charge moyenne |
| 7 à 9 | charge élevée |

Dans le projet, on commencera par la classification binaire.

#### Q27. Pourquoi commencer par une classification binaire plutôt qu’une classification ternaire ?

### Réponse

Certains individus sont incapables de distinguer les nuances à ce niveau de détail.
Une classification binaire est plus discriminante donc on réduit le potentiel de faux positifs et faux négatifs.

Exemple: Sur un score de 3, c'est plus facile de qualifier cela comme charge faible dans une échelle binaire que dans une échelle ternaire car la différence avec la classe la plus proche est plus grande.

## 9. Visualisation des signaux EEG

Avant tout prétraitement ou apprentissage automatique, il est indispensable de visualiser les signaux EEG.

La visualisation permet de :

- vérifier que les données sont correctement chargées ;
- observer les amplitudes ;
- repérer les valeurs aberrantes ;
- détecter les ruptures ou données manquantes ;
- comparer les canaux ;
- repérer des artefacts évidents ;
- comparer qualitativement des segments de faible et forte charge cognitive.

### Algorithme conceptuel pour visualiser un segment EEG

Pour visualiser un segment EEG de 10 secondes :

1. choisir un sujet ;
2. choisir un scénario ;
3. choisir un segment ;
4. récupérer les 4 canaux EEG ;
5. calculer le nombre d’échantillons : $N = f_s \times T$ ;
6. construire l’axe temporel : $t[n] = n/f_s$ ;
7. tracer les quatre canaux en fonction du temps ;
8. ajouter le score PAAS associé au titre du graphique ;
9. commenter la qualité du signal.
